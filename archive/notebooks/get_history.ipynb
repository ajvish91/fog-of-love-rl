# FoL run history and artifacts

Loads exports from **`archive/artifacts/`** (not tracked in Git). Place **`pareto_history.json`** there after downloading from W&B. Training runs via **`python -m fol.training.run_fol`** — see **`docs/training-cli.md`** and **`README.md`**.

**Weights & Biases.** Logged runs typically use projects such as `fol-local-abrl-agile-implement`, `fol-multi-local-abrl-agile-implement`, `fol-abrl-agile-implement`, or `fol-agile-implement` depending on mode and whether ABRL is enabled—check `scripts/fol/training/run_fol.py` for the exact mapping.

**Rich test traces** (actions / observations per test episode) are emitted for the **`arg_main`** configuration when not in debug mode. To reproduce a JSON file locally, export the relevant keys from a W&B run or point the cells below at your download path.

```python
from pathlib import Path

def _artifacts_dir() -> Path:
    cwd = Path.cwd()
    if (cwd / "archive" / "artifacts").is_dir():
        return (cwd / "archive" / "artifacts").resolve()
    if cwd.name == "notebooks" and (cwd.parent / "artifacts").is_dir():
        return (cwd.parent / "artifacts").resolve()
    return (cwd / "archive" / "artifacts").resolve()

ARTIFACTS = _artifacts_dir()
ARTIFACTS.mkdir(parents=True, exist_ok=True)

import json
```

```python
player_1_obs_history = []
player_1_act_history = []
player_2_act_history = []
player_2_obs_history = []
for episode_act, episode_obs in zip(data["test_actions"], data["test_observations"]):
    for step_act, step_obs in zip(episode_act, episode_obs):
        player_1_act_history.append(step_act['player_1'])
        player_2_act_history.append(step_act['player_2'])
        # Extract raw data for player_1 and player_2
        raw_data_player_1 = step_obs['player_1']
        raw_data_player_2 = step_obs['player_2']
        player_1_obs = [item for sublist in raw_data_player_1.values() for item in sublist[0]]
        player_2_obs = [item for sublist in raw_data_player_2.values() for item in sublist[0]]
        player_1_obs_history.append(player_1_obs)
        player_2_obs_history.append(player_2_obs)
player_1_act_history, player_1_obs_history
```

```python
player_1_obs_history_new = []
for obs in player_1_obs_history:
    obs_dict = {}
    for item, key in zip(obs, observation_space["player_1"]):
        obs_dict[key] = item
    player_1_obs_history_new.append(obs_dict)
len(player_1_obs_history_new)
        
```

```python
player_2_obs_history_new = []
for obs in player_2_obs_history:
    obs_dict = {}
    for item, key in zip(obs, observation_space["player_2"]):
        obs_dict[key] = item
    player_2_obs_history_new.append(obs_dict)
len(player_2_obs_history_new)
```

```python
import pickle
# Save to pickle file
with open(ARTIFACTS / 'player_1_obs.pkl', 'wb') as file:
    pickle.dump(player_1_obs_history_new, file)
    
with open(ARTIFACTS / 'player_2_obs.pkl', 'wb') as file:
    pickle.dump(player_2_obs_history_new, file)
    
with open(ARTIFACTS / 'player_1_act.pkl', 'wb') as file:
    pickle.dump(player_1_act_history, file)
    
with open(ARTIFACTS / 'player_2_act.pkl', 'wb') as file:
    pickle.dump(player_2_act_history, file)
```
